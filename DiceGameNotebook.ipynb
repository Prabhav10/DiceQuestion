```python
# Importing Packages
import random
```

```python
# Variables
trials = 100000
```

```python
# Level 1 Game 
dice = 6
target = 10
table = {}
for i in range(trials):
    # First Roll
    first_roll = random.randint(1, dice)
    
    # Following Game Rules
    if first_roll <= target - dice:
        second_roll = random.randint(1, dice)
        result = first_roll + second_roll
    else:
        result = first_roll 
    
    # Storing Outcomes in Dictionary
    if result not in table:
        table[result] = 0
    table[result] += 1
    
# Calculating Expected Value and Comparing it to the answer
ev = sum(map(lambda x: x[0] * x[1], table.items())) / trials
answer = 35 / 6
print(f'Answer: {round(answer, 2)} | Simulated Expected Value: {round(ev, 2)}')
```

```python
# Level 2 Game 
for size in range(1, 11):
    table = {}
    for _ in range(trials):
        first_roll = random.randint(1, size)
        if first_roll <= target - size:
            second_roll = random.randint(1, size)
            result = first_roll + second_roll
        else:
            result = first_roll 
        if result not in table:
            table[result] = 0
        table[result] += 1
    
    ev = sum(map(lambda x: x[0] * x[1], table.items())) / trials
    if 1 <= size <= 5:
        answer = size + 1
    else:
        answer = 5 + 5 / size
    print('{:<9s}{:<15s}{:<10s}'.format(f'n = {size}', f'Answer: {round(answer, 2)}', f'Simulated Expected Value: {round(ev, 2)}'))
```

```python
# Level 3 Game 
for target in range(1, 21):
    print(f'Target Sum: {target}')
    print('---------------------')
    for size in range(1, 11):
        table = {}
        for _ in range(trials):
            first_roll = random.randint(1, size)
            if first_roll <= target - size:
                second_roll = random.randint(1, size)
                result = first_roll + second_roll
            else:
                result = first_roll 
            if result not in table:
                table[result] = 0
            table[result] += 1

        ev = sum(map(lambda x: x[0] * x[1], table.items())) / trials
        if 1 <= target <= size:
            answer = (1 + size) / 2
        elif size + 1 <= target <= 2*size - 1:
            answer = (size + 1) * target / (2 * size)
        else:
            answer = 1 + size
        print('{:<9s}{:<15s}{:<10s}'.format(f'n = {size}', f'Answer: {round(answer, 2)}', f'Simulated Expected Value: {round(ev, 2)}'))
    print()
```

```python
# Level 4 Game

results = [] 
for n_1 in range(1, 11):
    for n_2 in range(1, 11):
        # Resetting Dictionary
        table = {}
        for _ in range(trials):
            # First Roll
            first_roll = random.randint(1, n_1)
            
            # Following Game Rules
            if first_roll <= target - n_2:
                second_roll = random.randint(1, n_2)
                result = first_roll + second_roll
            else:
                result = first_roll 
            
            # Storing Outcomes in Dictionary
            if (n_1, n_2, result) not in table:
                table[(n_1, n_2, result)] = 0
            table[(n_1, n_2, result)] += 1
        
        # Calculating Expected Value
        ev = sum(map(lambda x: x[0][-1] * x[1], table.items())) / trials
        
        # Calculating Theoretical Answer
        if n_1 + n_2 <= 10:
            answer = [(n_1, n_2, (n_1 + n_2 + 2) / 2, ev)]
        elif n_1 <= 10 and n_2 <= 10:
            answer = [(n_1, n_2, (n_1**2 - n_2**2 + n_1 + 9*n_2 + 10) / (2*n_1), ev)]
        elif n_2 >= 11:
            answer = [(n_1, n_2,(1 + n_1) / 2, ev)]
        
        # Adding results to an existing list
        results.extend(answer)


# Sorting list by expected value
results.sort(key=lambda x: x[2])

# Outputting results
for n_1, n_2, answer, ev in results:
    print('{:<15s}{:<15s}{:<15s}{:<15s}'.format(f'n_1 = {n_1}', 
                                         f'n_2 = {n_2}', 
                                         f'Answer: {round(answer, 2)}',
                                         f'Simulated Expected Value: {round(ev, 2)}'))
```

```python
# Level 5 Game

for target in range(1, 11):
    # Resetting Results
    results = [] 

    print(f'Target Sum: {target}')
    print('---------------------')
    
    for n_1 in range(1, target+1):
        for n_2 in range(1, target+1):
            # Resetting Dictionary
            table = {}
            for _ in range(trials):
                # First Roll
                first_roll = random.randint(1, n_1)

                # Following Game Rules
                if first_roll <= target - n_2:
                    second_roll = random.randint(1, n_2)
                    result = first_roll + second_roll
                else:
                    result = first_roll 

                # Storing Outcomes in Dictionary
                if (n_1, n_2, result) not in table:
                    table[(n_1, n_2, result)] = 0
                table[(n_1, n_2, result)] += 1

            # Calculating Expected Value
            ev = sum(map(lambda x: x[0][-1] * x[1], table.items())) / trials

            # Calculating Theoretical Answer
            if n_1 + n_2 <= target:
                answer = [(n_1, n_2, (n_1 + n_2 + 2) / 2, ev)]
            elif n_1 <= target and n_2 <= target:
                answer = [(n_1, n_2, (n_1**2 - n_2**2 + n_1 + (target-1)*n_2 + target) / (2*n_1), ev)]
            elif n_2 >= 11:
                answer = [(n_1, n_2,(1 + n_1) / 2, ev)]

            # Adding results to an existing list
            results.extend(answer)


    # Sorting list by expected value
    results.sort(key=lambda x: x[2])

    # Outputting results
    for n_1, n_2, answer, ev in results:
        print('{:<11s}{:<11s}{:<15s}{:<15s}'.format(f'n_1 = {n_1}', f'n_2 = {n_2}', f'Answer: {round(answer, 2)}', f'Simulated Expected Value: {round(ev, 2)}'))
    print()
```
